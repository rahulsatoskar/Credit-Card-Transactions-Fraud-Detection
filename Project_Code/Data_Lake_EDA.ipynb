```python
train_df['is_fraud'].value_counts()
```

```python
train_non_fraud_count = train_df['is_fraud'].value_counts()[0]
train_fraud_count = train_df['is_fraud'].value_counts()[1]
total_count = train_non_fraud_count + train_fraud_count
percent_train_non_fraud = (train_non_fraud_count/total_count) * 100
percent_train_fraud = (train_fraud_count/total_count) * 100
print("Percentage of Non Fraud transactions in train dataset is ", percent_train_non_fraud, "%")
print("\n")
print("Percentage of Fraud transactions in train dataset is ", percent_train_fraud, "%")
```

# Percentage of Non Fraud transactions in train dataset is  99.42113482561166%


# Percentage of Fraud transactions in train dataset is  0.5788651743883394%


# Note- We have a severe imbalanced train dataset. Oversampling/Undersampling techniques will be used to counter this.

```python
test_non_fraud_count = test_df['is_fraud'].value_counts()[0]
test_fraud_count = test_df['is_fraud'].value_counts()[1]
total_count = test_non_fraud_count + test_fraud_count
percent_test_non_fraud = (test_non_fraud_count/total_count) * 100
percent_test_fraud = (test_fraud_count/total_count) * 100
print("Percentage of Non Fraud transactions in test dataset is ", percent_test_non_fraud, "%")
print("\n")
print("Percentage of Fraud transactions in test dataset is ", percent_test_fraud, "%")
```
